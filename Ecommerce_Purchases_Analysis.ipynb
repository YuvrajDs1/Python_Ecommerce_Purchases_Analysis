# **How many people have a credit card that expires at 2020, 2025 and 2026**

```python
df.columns
```

```python
df['CC Exp Date']
```

```python
def function():
  count = 0
  for date in df['CC Exp Date']:
    if date.split('/')[1] == '20':
      count=count+1
  print(count)
```

```python
function()
```

```python
def TotalExp():
  count = 0
  for date in df['CC Exp Date']:
    if date.split('/')[1] == '25':
      count+=1
  print(count)
```

```python
TotalExp()
```

```python
len(df[df['CC Exp Date'].apply(lambda x:x[3:] == '26')])
```
